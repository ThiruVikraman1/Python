```python
#Create a function to check weather the given number is odd or even
def OddEven():
    num = int(input("Enter a number"))
    if num%2 == 0:
        print (num,"is even number")
    else:
        print (num,"is odd number")
```

```python
#Create a function that tells the eligiblity for marriage for male and female according to their age limit like 21 for male and 18 for female
def Eligible():
    Gender = input("Enter yot gender:")
    Age = int(input("Enter your Age"))
    print('Gender:',Gender)
    print('Age:',Age)
    if (Gender == 'male' and Age >=21):
        print("ELIGIBLE")
    elif (Gender == 'female' and Age>=18):
        print('ELIGIBLE')
    else:
        print("NOT ELIGIBLE")
```

```python
Eligible()
```

```python
#Calculate the prdentage of your 10th mark
def percentage():
    subject1 =int(input("Enter subject1 mark:"))
    subject2 =int(input("Enter subject2 mark:"))
    subject3 =int(input("Enter subject3 mark:"))
    subject4 =int(input("Enter subject4 mark:"))
    subject5 =int(input("Enter subject5 mark:"))
    total = (subject1+subject2+subject3+subject4+subject5)
    percentage = (total/500)*100
    return percentage
```

```python
percentage = percentage()
print("percentage:",percentage)
```

```python
#Print he area and perimeter of a triangle using function
def Triangle():
    height= int(input(" Enter Height:"))
    base = int(input("Enter Base"))
    print("Area formula: (Height*Base)/2")
    Area= (height*base)/2
    print("Area of triangle:",Area)
    side1=int(input("Side1:"))
    side2=int(input("Side2:"))
    base=int(input("Base:"))
    print("Perimeter formula: Side1+Side2+Base")
    Perimeter= side1+side2+base        
    print("Perimeter of Triangle:",Perimeter)
```
