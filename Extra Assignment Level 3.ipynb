```python
#Categorize the people by their age children, adult, citizen, senior citizen
age = int(input("enter your age:"))
if age < 18:
    print("Children")
elif age <35:
    print("Adult")
elif age < 59:
    print("Citizen")
else:
    print("Senior Citizen")
```
